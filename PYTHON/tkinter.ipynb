```python
import tkinter as tk

# Dictionary containing course names and their respective teachers
courses_teachers = {
    'Mathematics': 'Dr. John Smith',
    'Algorithms': 'Prof. Jane Doe',
    'Data Structures': 'Dr. Alan Turing',
    'Operating Systems': 'Prof. Grace Hopper',
    'Database Systems': 'Dr. Edgar Codd',
    'Software Engineering': 'Prof. Barbara Liskov',
    'Artificial Intelligence': 'Dr. Andrew Ng',
    'Machine Learning': 'Prof. Tom Mitchell',
    'Computer Networks': 'Dr. Leonard Kleinrock',
    'Cyber Security': 'Prof. Bruce Schneier',
}

# Creating the main window
root = tk.Tk()
root.title("MCA Courses and Teachers")

# Creating the listbox with scrollbars
frame = tk.Frame(root)
frame.pack(padx=10, pady=10)

vscrollbar = tk.Scrollbar(frame, orient=tk.VERTICAL)
vscrollbar.pack(side=tk.RIGHT, fill=tk.Y)

hscrollbar = tk.Scrollbar(frame, orient=tk.HORIZONTAL)
hscrollbar.pack(side=tk.BOTTOM, fill=tk.X)

listbox = tk.Listbox(frame, yscrollcommand=vscrollbar.set, xscrollcommand=hscrollbar.set, width=40, height=10)
listbox.pack(side=tk.LEFT)

for course in courses_teachers:
    listbox.insert(tk.END, course)

vscrollbar.config(command=listbox.yview)
hscrollbar.config(command=listbox.xview)

# Variable to hold the teacher's name
teacher_name_var = tk.StringVar()

# Creating a label and text box to display the teacher's name
label = tk.Label(root, text="Subject Teacher:")
label.pack(pady=(10, 0))

teacher_name_entry = tk.Entry(root, textvariable=teacher_name_var, state='readonly', width=50)
teacher_name_entry.pack(pady=(0, 10))

# Bind the listbox selection event directly
listbox.bind('<<ListboxSelect>>', lambda event: teacher_name_var.set(courses_teachers[listbox.get(listbox.curselection())]))

# Running the main event loop
root.mainloop()
```

Button : a button is basically a label that performs event handling.
a button can perform left click event an rightclick event separately. After event execution the process that takes place is known as action.

```python
from tkinter import *
win = Tk()
win.title("Python GUI App")

lbl = Label(win, text="Button not clicked", foreground = 'red')
lbl.pack()

def click():
    lbl.configure(text="Button clicked")
    lbl.configure(foreground='green')
    action.configure(text="Clicked")

action = Button(win, text="Click me", background='red', command=click)
action.pack()

win.mainloop()
```

```python
from tkinter import *

# Prices for the menu items
prices = {
    "Poha": 20,
    "Sandwich": 50,
    "Chapati": 10,
    "Paratha": 15,
    "Dal": 30,
    "Curd": 20,
    "Rice": 25,
    "Roti": 10
}

# List to store selected items
selected_items = []

def on_item_select(item_name):
    selected_items.append(item_name)
    update_bill()

def update_bill():
    # Clear the text box
    text_box.delete(1.0, END)
    
    # Print selected items
    text_box.insert(END, "Selected Items:\n")
    total_price = 0
    for item in selected_items:
        text_box.insert(END, f"{item} - ${prices[item]}\n")
        total_price += prices[item]

    # Print total items count and total price
    text_box.insert(END, "\n")
    text_box.insert(END, f"Total Items: {len(selected_items)}\n")
    text_box.insert(END, f"Total Price: ${total_price}\n")

top = Tk()
top.title("Menu Selection Example with Bill Generator")

# Create a Text box to display selected menu items and bill
text_box = Text(top, height=20, width=40)
text_box.pack()

# Create a toplevel menu
menubar = Menu(top)

# Breakfast menu
breakfast = Menu(menubar, tearoff=0)
breakfast.add_command(label="Poha", command=lambda: on_item_select("Poha"))
breakfast.add_command(label="Sandwich", command=lambda: on_item_select("Sandwich"))

# Lunch menu
lunch = Menu(menubar, tearoff=0)
lunch.add_command(label="Chapati", command=lambda: on_item_select("Chapati"))
lunch.add_command(label="Paratha", command=lambda: on_item_select("Paratha"))
lunch.add_command(label="Dal", command=lambda: on_item_select("Dal"))
lunch.add_command(label="Curd", command=lambda: on_item_select("Curd"))

# Dinner menu
dinner = Menu(menubar, tearoff=0)
dinner.add_command(label="Rice", command=lambda: on_item_select("Rice"))
dinner.add_command(label="Roti", command=lambda: on_item_select("Roti"))
dinner.add_command(label="Dal", command=lambda: on_item_select("Dal"))

# Add the menus to the menubar
menubar.add_cascade(label="Breakfast", menu=breakfast)
menubar.add_cascade(label="Lunch", menu=lunch)
menubar.add_cascade(label="Dinner", menu=dinner)

# Display the menu
top.config(menu=menubar)

# Run the main loop
top.mainloop()
```

```python
from tkinter import *
from fpdf import FPDF
import datetime

# Prices for the menu items
prices = {
    "Poha": 20,
    "Sandwich": 50,
    "Chapati": 10,
    "Paratha": 15,
    "Dal": 30,
    "Curd": 20,
    "Rice": 25,
    "Roti": 10
}

# List to store selected items
selected_items = []

def on_item_select(item_name):
    selected_items.append(item_name)
    update_bill()

def update_bill():
    # Clear the text box
    text_box.delete(1.0, END)
    
    # Print selected items
    text_box.insert(END, "Selected Items:\n")
    total_price = 0
    for item in selected_items:
        text_box.insert(END, f"{item} - ${prices[item]}\n")
        total_price += prices[item]

    # Print total items count and total price
    text_box.insert(END, "\n")
    text_box.insert(END, f"Total Items: {len(selected_items)}\n")
    text_box.insert(END, f"Total Price: ${total_price}\n")

def generate_pdf():
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    
    # Add title and date
    pdf.cell(200, 10, txt="Bill Receipt", ln=True, align='C')
    pdf.cell(200, 10, txt=str(datetime.datetime.now()), ln=True, align='C')
    pdf.ln(10)
    
    # Add selected items
    pdf.cell(200, 10, txt="Selected Items:", ln=True)
    total_price = 0
    for item in selected_items:
        pdf.cell(200, 10, txt=f"{item} - ${prices[item]}", ln=True)
        total_price += prices[item]
    
    # Add total count and price
    pdf.ln(10)
    pdf.cell(200, 10, txt=f"Total Items: {len(selected_items)}", ln=True)
    pdf.cell(200, 10, txt=f"Total Price: ${total_price}", ln=True)
    
    # Save the PDF
    pdf.output("Bill.pdf")
    print("Bill saved as Bill.pdf")

top = Tk()
top.title("Menu Selection Example with Bill Generator")

# Create a Text box to display selected menu items and bill
text_box = Text(top, height=20, width=40)
text_box.pack()

# Create a toplevel menu
menubar = Menu(top)

# Breakfast menu
breakfast = Menu(menubar, tearoff=0)
breakfast.add_command(label="Poha", command=lambda: on_item_select("Poha"))
breakfast.add_command(label="Sandwich", command=lambda: on_item_select("Sandwich"))

# Lunch menu
lunch = Menu(menubar, tearoff=0)
lunch.add_command(label="Chapati", command=lambda: on_item_select("Chapati"))
lunch.add_command(label="Paratha", command=lambda: on_item_select("Paratha"))
lunch.add_command(label="Dal", command=lambda: on_item_select("Dal"))
lunch.add_command(label="Curd", command=lambda: on_item_select("Curd"))

# Dinner menu
dinner = Menu(menubar, tearoff=0)
dinner.add_command(label="Rice", command=lambda: on_item_select("Rice"))
dinner.add_command(label="Roti", command=lambda: on_item_select("Roti"))
dinner.add_command(label="Dal", command=lambda: on_item_select("Dal"))

# Add the menus to the menubar
menubar.add_cascade(label="Breakfast", menu=breakfast)
menubar.add_cascade(label="Lunch", menu=lunch)
menubar.add_cascade(label="Dinner", menu=dinner)

# Display the menu
top.config(menu=menubar)

# Add a button to generate PDF
pdf_button = Button(top, text="Print Bill to PDF", command=generate_pdf)
pdf_button.pack()

# Run the main loop
top.mainloop()



```

Treeview() - to create a table
the tree view function takes minimum two arguments that is rows and columns. These rows and columns must display on some canvas and that canvas is always the very first argument.
    once the table is created we can mention the column id and the column title. The ids are created with the function name - column().
To provide the title of column the function name is heading(). it takes minimum two arguments - columnid and the text that to be displayed as a title.
Once the table structureis created with:
1. treeview
2. columnid
3. columntitle
   Then we can insert the value inside the table. To insert the value we required insert command. this command inserts a single row at a time that means we have to provide all the column values together in a single insert command. The parameters of this command are as follows:
   1. parent -
   2. Index - this parameter asks for the last value which should be included in a row. If we want to provide all the values for all the columns which present in the table, we have to mention "end".
   3. IIE(index identifier): it is used to provide an index number for the rows what ever we are inserting. For the inde identifier if we want to privide any name to that row, we can mwention the text value as a next parameter.
   4. Values -- This parameter contains values for all the columns separated by comma(,). 

```python
from tkinter import *
from tkinter import ttk

root = Tk()

# Create Treeview with columns
tab = ttk.Treeview(root, columns=("0", "1", "2", "3"))

# Creating column headings
tab.column("#0", width=50)
tab.heading("#0", text="Sl.No")
tab.column("#1", width=100)
tab.heading("#1", text="Regis.No")
tab.column("#2", width=100)
tab.heading("#2", text="Student Name")
tab.column("#3", width=100)
tab.heading("#3", text="Result")

# Inserting column values
tab.insert(parent='', index='end', iid=1, text="1", values=('101', 'ABC', 'Pass'))
tab.insert(parent='', index='end', iid=2, text="2", values=('102', 'XYZ', 'Fail'))

tab.pack()

root.mainloop()
```
