Class: 
1. a class is a blueprint or the template that combines member data and the member functions.
2. a class is not an executable unit, to execute the member functions of a class it is required to be having some instances.
3. a class provides the facility of encapsulation, where multiple data members and the functions are binded together and remain hidden from the outside    world.
4. a class doesn't take space on the memory, because it is not an executable unit. the memory space is occupied by the objects of a class.
5. a class is a user defined data type and it is created with class keyword.
6. a class can be an abstract class that doesn't have any functional body that is known as abstract class.
7. a class can be inherited by another class.

Object:
1. An object is an instance of a class.
2. An object represents a single class at a time while a single class can have multilobjects at a time.
3. An object can assess data and the funvctions of the class.
4. An object is an executable for a class.
5. The object of a same class can execute any of the function present in the class.
6. objects can't be inhrited, they can't show the polymorphism, encasulation and the abstraction.
7. objects can be created during run time, that is known as real world object creation.
8. for an API the objects represents the interfaces.


Polymorphism:
there are two types of polymorphism
   * compiletime polymorphism - method overloading => uses function of same name more than once.
   rules:
   the datatype of the argument must be different.
   the number of arguments must be diferent.
   sequence of argument must be different.
   the return type of the function must be different.

   * run time polymorphism: it can be implemented with the concept of inheritance.
   in this we have method over riding - spetial definition to the function to the child class which is already present in the parent class.

Inheritance:
1. When one class calls the propertes of anther class then the concept is known as inheritance.
2. the class that provides its properties called parent class.
3. the class that inherits the properties is called child class.
4. a child class has its own property including the adopted property.

* the inheritance executes properly with the object of the child class. A parent class object can work parallerly wothout calling the properties of the child class.

* there may be a possibility the parent class and the chind classes have same name of functions, in this sutuation the child class functions takes priority to get execute and they override the functionality of parent class.

Inheritance:
1. single
2. multiple
3. hierarchical
4. hybrid
5. multi level

Abstractions:
1. it is a property to hiding the data values. abstraction can be done with the help of abstract classes and the abstract functions. a class with out function body is known as abstract class and a function without a body is known as anstract function.
2. abstract functions are the speciality of abstract classes.

```python
class Vehicle:
    def set_num_wheels(self, num_wheels):
        self.num_wheels = num_wheels

    def set_engine_type(self, engine_type):
        self.engine_type = engine_type

    def set_engine_number(self, engine_number):
        self.engine_number = engine_number

    def get_num_wheels(self):
        return self.num_wheels

    def get_engine_type(self):
        return self.engine_type

    def get_engine_number(self):
        return self.engine_number

    def display_info(self):
        print(f"Number of wheels: {self.get_num_wheels()}")
        print(f"Engine type: {self.get_engine_type()}")
        print(f"Engine number: {self.get_engine_number()}")

    def is_two_wheeler(self):
        return self.get_num_wheels() == 2

    def is_three_wheeler(self):
        return self.get_num_wheels() == 3

# Create an object for a two-wheeler
two_wheeler = Vehicle()
two_wheeler.set_num_wheels(2)
two_wheeler.set_engine_type("Single-cylinder")
two_wheeler.set_engine_number("TW123456")

# Create an object for a three-wheeler
three_wheeler = Vehicle()
three_wheeler.set_num_wheels(3)
three_wheeler.set_engine_type("Twin-cylinder")
three_wheeler.set_engine_number("TH789012")

# Execute various functions with the two-wheeler object
print("Two-Wheeler Info:")
two_wheeler.display_info()
print(f"Is it a two-wheeler? {two_wheeler.is_two_wheeler()}")
print(f"Is it a three-wheeler? {two_wheeler.is_three_wheeler()}")

print("\n-----------------------------\n")

# Execute various functions with the three-wheeler object
print("Three-Wheeler Info:")
three_wheeler.display_info()
print(f"Is it a two-wheeler? {three_wheeler.is_two_wheeler()}")
print(f"Is it a three-wheeler? {three_wheeler.is_three_wheeler()}")
```
